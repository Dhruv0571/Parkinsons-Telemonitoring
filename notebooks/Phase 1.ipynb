```python
relevant_feature = 'Jitter(%)'

# Minimum and Maximum
min_value = df[relevant_feature].min()
max_value = df[relevant_feature].max()

# Mean, Median, and Mode
mean_value = df[relevant_feature].mean()
median_value = df[relevant_feature].median()
mode_value = df[relevant_feature].mode()[0]

# Range
range_value = max_value - min_value

# Variance and Standard Deviation
variance_value = df[relevant_feature].var()
sd_value = df[relevant_feature].std()

# 1st, 2nd, and 3rd Quartiles
q1 = df[relevant_feature].quantile(0.25)
q2 = df[relevant_feature].quantile(0.5)
q3 = df[relevant_feature].quantile(0.75)
```

#### Data Exploration Task 4

```python
def FrequencyDistribution(d):
    F = pd.cut(d, bins=10)
    FRC = {}
    CRF = 0; RF = 0
    for i in F:  
        mid_point = (i.left + i.right) / 2
        RF = i.right / len(d)
        CRF += RF
        FRC[mid_point] = [RF, CRF]
    return FRC

fd = FrequencyDistribution(df[relevant_feature])
```

```python
print(f"Frequency Distribution: {fd}")
```

#### Data Exploration Task 5

```python
def FrequencyDistribution_mean(d):
    F = FrequencyDistribution(d)
    N = 0; D = 0;
    for mid_point, values in F.items():
        RF, CRF = values
        N += mid_point * RF
        D += RF
    Fx = N / D
    return Fx

fdm = FrequencyDistribution_mean(df[relevant_feature])
print(f"Mean of Frequency Distribution: {fdm:.2f}")
```

#### Data Shaping Task 2

Table: Patients

Columns: patient_id, subject_id, age, sex
Constraints: 
patient_id: PRIMARY KEY, AUTO_INCREMENT: This column serves as the primary key and is auto-incremented. It uniquely identifies each patient record.

subject_id: This column presumably stores the subject ID associated with each patient.

age: Stores the age of the patients.

sex: This column likely represents the gender of the patients. It's a numerical representation (e.g., 0 for Male, 1 for Female).

Table: Tests

Columns: test_id, patient_id, test_time, total_UPDRS
Constraints:
test_id: PRIMARY KEY, AUTO_INCREMENT: This column serves as the primary key and is auto-incremented. It uniquely identifies each test record.

patient_id: FOREIGN KEY: This column references the patient_id in the Patients table. It establishes a relationship between the Tests table and the Patients table, ensuring that each test is associated with a patient.

Table: Results

Columns: result_id, test_id, Jitter_percent, Jitter_Abs, Jitter_RAP, Jitter_PPQ5, Jitter_DDP, Shimmer, Shimmer_dB, Shimmer_APQ3, Shimmer_APQ5, Shimmer_APQ11, Shimmer_DDA

Constraints:

result_id: PRIMARY KEY, AUTO_INCREMENT: This column serves as the primary key and is auto-incremented. It uniquely identifies each result record.

test_id: FOREIGN KEY: This column references the test_id in the Tests table. It establishes a relationship between the Results table and the Tests table, ensuring that each result is associated with a test.

Relationships:

Patients to Tests: The Patients table has a one-to-many relationship with the Tests table. One patient can have multiple tests. This relationship is established using the patient_id column in the Tests table, which references the patient_id in the Patients table.

Tests to Results: The Tests table has a one-to-many relationship with the Results table. One test can have multiple results. This relationship is established using the test_id column in the Results table, which references the test_id in the Tests table.

```python
conn = mysql.connector.connect(
    user='root',
    password='YOUR_PASSWORD_HERE',  # Replace with your actual password
    host='localhost',
    database='ParkinsonsDB'
)

cursor = conn.cursor()


cursor.execute("CREATE DATABASE IF NOT EXISTS ParkinsonsDB")

cursor.execute("USE ParkinsonsDB")

# Create Patients table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS Patients (
        patient_id INT AUTO_INCREMENT PRIMARY KEY,
        subject_id INT,
        age INT,
        sex VARCHAR(10)
    )
""")

cursor.execute("""
    ALTER TABLE Patients
    MODIFY COLUMN sex INTEGER
""")

# Create Tests table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS Tests (
        test_id INT AUTO_INCREMENT PRIMARY KEY,
        patient_id INT,
        test_time FLOAT,
        total_UPDRS FLOAT,
        FOREIGN KEY (patient_id) REFERENCES Patients(patient_id)
    )
""")

# Create Results table 
cursor.execute("""
    CREATE TABLE IF NOT EXISTS Results (
        result_id INT AUTO_INCREMENT PRIMARY KEY,
        test_id INT,
        Jitter_percent FLOAT,
        Jitter_Abs FLOAT,
        Jitter_RAP FLOAT,
        Jitter_PPQ5 FLOAT,
        Jitter_DDP FLOAT,
        Shimmer FLOAT,
        Shimmer_dB FLOAT,
        Shimmer_APQ3 FLOAT,
        Shimmer_APQ5 FLOAT,
        Shimmer_APQ11 FLOAT,
        Shimmer_DDA FLOAT,
        FOREIGN KEY (test_id) REFERENCES Tests(test_id)
    )
""")

conn.commit()
conn.close()
```

```python
conn = mysql.connector.connect(
    user='root',
    password='YOUR_PASSWORD_HERE',  # Replace with your actual password
    host='localhost',
    database='ParkinsonsDB'
)

cursor = conn.cursor()
inserted_subject_ids = set()

for index, row in df.iterrows():
    subject_id = row['subject#']
    if subject_id not in inserted_subject_ids:
        cursor.execute("""
            INSERT INTO Patients (subject_id, age, sex)
            VALUES (%s, %s, %s)
        """, (subject_id, row['age'], row['sex']))
        inserted_subject_ids.add(subject_id)

for index, row in df.iterrows():
    cursor.execute("""
        INSERT INTO Tests (patient_id, test_time, total_UPDRS)
        VALUES ((SELECT patient_id FROM Patients WHERE subject_id = %s Limit 1), %s, %s)
    """, (row['subject#'], row['test_time'], row['total_UPDRS']))

for index, row in df.iterrows():
    cursor.execute("""
        INSERT INTO Results (test_id, Jitter_percent, Jitter_Abs, Jitter_RAP, Jitter_PPQ5, Jitter_DDP, Shimmer, Shimmer_dB, Shimmer_APQ3, Shimmer_APQ5, Shimmer_APQ11, Shimmer_DDA)
        SELECT t.test_id, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
        FROM Tests t
        INNER JOIN Patients p ON t.patient_id = p.patient_id
        WHERE p.subject_id = %s
    """, (row['Jitter(%)'], row['Jitter(Abs)'], row['Jitter:RAP'], row['Jitter:PPQ5'], row['Jitter:DDP'], row['Shimmer'], row['Shimmer(dB)'], row['Shimmer:APQ3'], row['Shimmer:APQ5'], row['Shimmer:APQ11'], row['Shimmer:DDA'], row['subject#']))

conn.commit()
conn.close()
```
